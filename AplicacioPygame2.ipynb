```python
import pygame
import socket

pygame.init()
def fontsize2(size):
	font = pygame.font.SysFont("Arial", size)
	return font

font_default = fontsize2(15) # Set default font size

labels = []
class Label:

	''' CLASS FOR TEXT LABELS ON THE WIN SCREEN SURFACE '''
	def __init__(self, screen, text, x, y, size=20, color="white"):
		if size != 20: # If the size is not the default, change the font size
			self.font = fontsize2(size)
		else:
			self.font = font_default
		self.image = self.font.render(text, 1, color) # Create the surface of the text image
		_, _, w, h = self.image.get_rect() # Get the width and height of the rectangle covering the surface
		self.rect = pygame.Rect(x, y, w, h) # Create a rectangle with the width and height of the text
		self.screen = screen
		self.text = text
		labels.append(self)

	def change_text(self, newtext, color="white"):
		self.image = self.font.render(newtext, 1, color) # Create a new surface with the new text

	def change_font(self, font, size, color="white"):
		self.font = pygame.font.SysFont(font, size) # Change the font and size
		self.change_text(self.text, color)

	def draw(self):
		self.screen.blit(self.image, (self.rect)) # Draw the text on the screen


def show_labels():
	for _ in labels:
		_.draw() # Draw all the labels on the screen


if __name__ == '__main__' and False:
	# Set up the screen
	info = pygame.display.Info()
	screen_width = info.current_w*0.25
	screen_height = info.current_h*0.8
	WINDOW_SIZE = (screen_width, screen_height)
	win = pygame.display.set_mode((screen_width, screen_height))
	clock = pygame.time.Clock()

	second = Label(win, "vueling+", screen_width*0.19, screen_height*0.2, 24, color="white")
	second.change_font("Arial", 40, "white")
	# LOOP TO MAKE THINGS ON THE SCRREEN
	loop = 1
	while loop:
		win.fill(0) # CLEAN THE SCREEN EVERY FRAME
		# CODE TO CLOSE THE WINDOW
		for event in pygame.event.get():
			if event.type == pygame.QUIT:
				loop = 0
			if event.type == pygame.KEYDOWN:
				if event.key == pygame.K_ESCAPE:
					loop = 0
		# CODE TO SHOW TEXT EVERY FRAME
		show_labels()
		pygame.display.update()
		clock.tick(60)

	pygame.quit()
```
